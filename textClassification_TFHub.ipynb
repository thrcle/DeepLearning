#1. Load Data

```
imdb = keras.datasets.imdb # 케라스에 내장된 데이터셋 불러오기
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000) #num_word=문장 길이 제한
```

```
word_index = imdb.get_word_index()
```

```
word_index
```

```
word_index = {k:(v+3) for k,v in word_index.items()}  # 모든 val=val+3
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2  # unknown
word_index["<UNUSED>"] = 3
```

```
word_index
```

```
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])  # key-val을 바꾼 dic을 새로 생성

def decode_review(text):
    return ' '.join([reverse_word_index.get(i, '?') for i in text])
```

```
reverse_word_index
```

```
decode_review(train_data[0])
```

#2. Text Classification

```
train_data = keras.preprocessing.sequence.pad_sequences(train_data, value=word_index['<PAD>'], padding='post', maxlen=256)
test_data = keras.preprocessing.sequence.pad_sequences(test_data, value=word_index['<PAD>'], padding='post', maxlen=256)
```

```
len(train_data[0]), len(train_data[102])  # 218, 790 -> 256, 256 으로 동일해짐
```

```
train_data[0]   # 남는 부분은 0으로 padding됨. 부족한 부분은 잘림
```

```
vocab_size = 10000

model = keras.Sequential()
#
model.add(keras.layers.Embedding(vocab_size, 16, input_shape=(None,)))  #vocasize만큼 단어 임베딩 벡터라이징-> 연관관계 함축적으로 표현
model.add(keras.layers.GlobalAveragePooling1D())  #하나의 dimension에 대한 pooling
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))    # binary classification

model.summary()
```

```
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
# train의 일부를 vali로 쪼갬
particial_x_train = train_data[10000:]
x_val = train_data[:10000]

particial_y_train = train_labels[10000:]
y_val = train_labels[:10000]
```

```
history = model.fit(particial_x_train, particial_y_train, epochs=50, batch_size=512,
                    validation_data=(x_val, y_val))
```

```
results = model.evaluate(test_data, test_labels)
print(results)
```
